# Pivot table creation

- Convert long format → wide format.
- Structure:
    -  Rows → Book-Title
    -  Columns → User-ID
    -  Values → Book-Rating
- Each book becomes a vector in high-dimensional space

````python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
book_pivot
````

**Note**
- Replace `NAN` with `0`
- `NAN` ==> NotANumber

````python
book_pivot.fillna(0,inplace=True)
````

````python
book_pivot
````

**Note**
- our data contains most of `0`s ==> sparse data
- it will consumes lot of memory
- A CSR matrix (Compressed Sparse Row matrix) is a memory-efficient way to store sparse data
- It is heavily used in machine learning, recommender systems, NLP, and GenAI because real datasets (ratings, tokens, embeddings) are often sparse
- in scipy `csr_matrix` is present
- https://docs.scipy.org/doc/scipy/reference/generated/scipy.sparse.csr_matrix.html

````python
from scipy.sparse import csr_matrix
````

````python
# help(csr_matrix)
````

````python
import numpy as np
````

````python
book_sparse = csr_matrix(book_pivot)
````

````python
type(book_sparse)
````

```
data ==> CSR format data array of the matrix

indices ==> CSR format index array of the matrix

indptr ==> CSR format index pointer array of the matrix
```

````python
book_sparse.data
````

````python
book_sparse.indices
````

````python
book_sparse.indptr
````

````python
from sklearn.neighbors import NearestNeighbors
````

````python
model = NearestNeighbors(algorithm='brute')
````

````python
model.fit(book_sparse)
````

# Testing the model using a function

````python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
````

````python
recommend_book("2nd Chance")
````
